```python
#in the following we want to have a closer look at the Adj Close and Adj Close Change in % 
#to make it mor 
dax["Adj Close in Million"]=dax["Adj Close"]/1000000
dax.head()
```

```python
sns.lineplot(x="Date", y="Adj Close in Million", hue="Index", data=dax)
sns.set(rc={'figure.figsize':(30,14)})
```

```python
#as the month May 2019 until October 2019 are pretty solid, we want to focus on the last half year
dax_nov = dax[(dax['Date'] >= '2019-11-01')]
dax_nov.head()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Date", y="Adj Close in Million", hue="Index", data=dax_nov)
plt.title("Dax Indices Develeopment Nov 2019 - May 2020")
```

```python
#Plot daily Change (%) of Indices bwtween November and 
sns.lineplot(x="Date", y="Adj Close Change in %", hue="Index", data=dax_nov)
```

```python
dax_nov = dax_nov.drop(["Open", "High", "Low", "Close", "Adj Close", "Adj Close in Million"], axis=1)

dax_pivot_nov = dax_nov.pivot_table(index = "Date", columns = "Index", aggfunc = "sum")
dax_pivot_nov
```

```python
change_week_sum= dax_pivot_nov.resample("W").sum()
change_week_sum.plot()
```

```python
change_week_avg= dax_pivot_nov.resample("W").mean()
change_week_avg.plot()
```

```python
change_month_sum= dax_pivot_nov.resample("M").sum()

change_month_sum.plot(figsize=(12,7))
plt.title("Dax Indices - Monthly Chnage in %")
```

```python
change_month_avg= dax_pivot_nov.resample("M").mean()
change_month_avg.plot(figsize=(12,7))
plt.title("Dax Indices - Avgerage Daily Chnage per Month in %")
```
