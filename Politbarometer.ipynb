```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Politbarometer.csv")
df
```

```python
df = df.sort_index(ascending=False)
df
```

```python
fig, ax = plt.subplots(1, figsize= (10,10))
ax.set_xlabel("Datum")
ax.set_ylabel("Prozent")
ax.yaxis.grid(True, linestyle = "--")

plt.plot("Datum", "CDU/CSU", data=df, color ="black", linewidth = 2)
plt.plot("Datum", "SPD", data=df, color = "red", linewidth = 2 )
plt.plot("Datum", "AfD", data=df, color = "blue", linewidth = 2)
plt.plot("Datum", "FDP", data=df, color = "yellow", linewidth = 2)
plt.plot("Datum", "Linke", data=df, color = "magenta", linewidth = 2)
plt.plot("Datum", "Grüne", data=df, color = "green", linewidth = 2)
plt.plot("Datum", "Sonstige", data=df, color = "grey", linewidth = 2)

plt.legend(loc="upper left", title = "Parteien", frameon=True)
plt.title("Germany: 'Politbarometer' Development")

#Pie-Chart Januar|Mai I
```

```python
df_Jan = df.loc[[6,]]
df_Jan = df_Jan.drop(("Datum"), axis=1)
#df_Jan = df_Jan.T
df_Jan
```

```python
df_Mai = df.loc[[0,]]
df_Mai = df_Mai.drop(("Datum"), axis=1)
#df_Mai = df_Mai.T
df_Mai
```

```python
df_Jan.T.plot(kind="pie",
              subplots=True, 
              figsize=(8,8), 
              colors=["black", "red", "blue", "yellow", "magenta", "green", "grey"])
plt.legend(loc="upper right", title = "Parteien", bbox_to_anchor=(1.15, 1))
plt.axes().set_ylabel("Januar")
plt.title("Pie Charts: Survey Parties")

df_Mai.T.plot(kind="pie", 
              subplots=True, 
              figsize=(8,8), 
              colors=["black", "red", "blue", "yellow", "magenta", "green", "grey"])
plt.legend(loc="upper right", title = "Parteien", bbox_to_anchor=(1.15, 1))
plt.axes().set_ylabel("Mai")
#plt.title("Pie Charts: Survey Parties")
```
